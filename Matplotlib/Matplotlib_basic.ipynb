2. Hai cách để vẽ: Pyplot API và Object-Oriented API
* Pyplot API: Nhanh chóng, tiện lợi
* OO API: dùng cho các đồ thị nâng cao

Pyplot API

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "green")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

Thêm limit cho x và y, thêm linestyle cho đường đồ thị

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "green", linestyle = "dashed")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.xlim([0,4])
# plt.ylim([-0.5,0.5])

# or we can use 
# plt.axis([xmin, xmax, ymin, ymax]) :)) ghê hông
plt.show()
```

Vẽ hai đồ thị trên cùng một plot:

```python
x = np.linspace(0,10,1000)
plt.plot(x, np.sin(x), color = "green", linestyle = "--", label = "sin(x)") # -- or dashed
plt.plot(x, np.cos(x), color = "red", label = "cos(x)")

plt.title("sin(x) và cos(x)")
plt.xlabel("x")
plt.ylabel("y")
# show ra cái label của hai đồ thị
plt.legend()
plt.show()
```

Object-Oriented API

```python
x = [1,2,3,4]
y = [10,20,30,40]

fig, ax = plt.subplots(figsize = (5,5), facecolor="#81ecec")

ax.plot(x,y);
ax.set(title = "A simple plot",
       xlabel = "x",
       ylabel = "y");
```

Scatter

```python
x = np.random.randint(0,100,20)
y = np.random.randint(0,100,20)
```

```python
# Pyplot API
plt.scatter(x, y);
```

```python
# OOP API
fig, ax = plt.subplots()
ax.scatter(x, y)
```

Để mỗi chấm có kích thước khác nhau, thêm màu sắc, thêm độ mờ, thêm color bar

```python
fig, ax = plt.subplots()
sizes = np.random.randint(50,1000,20)
colors = np.random.randint(50,1000,20)
plot1 = ax.scatter(x, y, s = sizes, c= colors, cmap = "magma", alpha = 0.3);

fig.colorbar(plot1)
```

Bar chart
- Vertical
- Horizonal

```python
# Prepare data from python dictionary
heights = {
    "Trung" : 173,
    "Kiet": 165,
    "Nguyen": 175,
    "Duy": 175
}
fig, ax = plt.subplots()
ax.bar(heights.keys(), heights.values());
```

```python
# Prepare data from python dictionary
heights = {
    "Trung" : 173,
    "Kiet": 165,
    "Nguyen": 175,
    "Duy": 175
}
fig, ax = plt.subplots()
ax.barh(list(heights.keys()), list(heights.values()));
# Hàm barh hơi đặc biệt
```

Histogram

```python
# Prepare data
student_height = np.random.normal(170, 10, 200)
```

```python
# Pyplot API
plt.hist(student_height, bins = 30);
```

Subplots

- Option 1:

```python
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(
    nrows = 2,
    ncols = 2,
    figsize = (10,5)
)

ax1.scatter(x, y);
ax2.hist(student_height);
ax3.plot(x, x **2);
```

- Option 2:

```python
fig, ax = plt.subplots(
    nrows = 2,
    ncols = 2,
    figsize = (10,5)
)

ax[0,0].scatter(x, y);
ax[0,1].hist(student_height);
ax[1,0].plot(x, x **2);
```
