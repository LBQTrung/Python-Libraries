3. loc và iloc (location và index location)

-> Giúp access các phần tử với các điều kiện nhất định

```python
df.head()
# Mặc định ra 5 row đầu tiên
```

* loc:

Muốn lọc ra các phần từ có quantity là 15:

```python
df.loc[df.quantity == 15]
# Có 2 tham số nếu truyền 1 tham số thì là lọc
```

Muốn lọc ra các phần từ có quantity là 15 hoặc item_name là Nantucket Nectar:

```python
df.loc[(df.item_name == "Nantucket Nectar") | (df.quantity == 15)]
# Chú ý: từng điều kiện
```

Muốn lọc ra các phần từ có quantity là 2 và item_name là Nantucket Nectar:

```python
df.loc[(df.quantity == 2) & (df.item_name == "Nantucket Nectar")]
```

Đối số thứ hai chính là những cái cột mà mình muốn chú tâm vào:

```python
df.loc[(df.quantity == 2) & (df.item_name == "Nantucket Nectar"), ["order_id", "quantity", "item_name"]]
```

4. Data Manipulation (Thao tác với dữ liệu)

Xem kiểu dữ liệu của một cột

```python
df.item_price.dtype
# Object đó :)) O buồi
```

Muốn chuyển các phần từ của item_price về dạng float ta làm như sau:

```python
df.item_price.apply(lambda x : float(x[1:]))
```

Vì chúng ta chưa gán trở lại item_price nên trong dataframe vẫn chưa có gì thay đổi:

```python
df.head()
```

Chúng ta giải quyết như sau:

```python
df.item_price = df.item_price.apply(lambda x : float(x[1:]))
df.head()
```

Tạo thêm một column mới là total_price bằng quantity nhân với item_price

```python
df["total_price"] = df["quantity"] * df["item_price"]
# Lưu ý dạng df. chỉ dành cho các column đã có 
# Để tạo mới ta dùng dạng ở trên
df.head()
```

Để tính tổng doanh thu:

```python
revenue = df.total_price.sum()
revenue
```

Tổng số lượng tương ứng với mỗi item_name ta dùng:

Group by

```python
df.groupby("item_name")["quantity"].sum()
```

5. Sắp xếp

* Sắp xếp tăng dần:

```python
c = df.groupby("item_name")["quantity"].sum()
c.sort_values()
```

* Sắp xếp giảm dần và lấy 5 phần từ tổng quantity nhiều nhất:

```python
c.sort_values(ascending = False).head()
```
