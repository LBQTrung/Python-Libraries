# PIL là viết tắt của  Python Imaging Library, là một thư viện với khả năng xử lí ảnh khá mạnh mẽ
  

import các thư viện cần thiết:

```python
import PIL
import numpy as np
```

Đọc bức ảnh vào bộ nhớ:

```python
origin_img = PIL.Image.open("yasuo.jpg")
```

In bức ảnh ra màn hình:

```python
origin_img
```

Chuyển bức ảnh từ cấu trúc dữ liệu của PIL sang Numpy array:

```python
img_array = np.array(origin_img.getdata())
```

Xuất số chiều của array kết quả:

```python
img_array.ndim
```

Xuất kích thước của array kết quả:

```python
img_array.shape
```

In nội dung của array kết quả:

```python
print(img_array)
```

Một vài thao tác với ảnh:

* Chuyển ảnh màu sang ảnh xám:

```python
gray_img = origin_img.convert("L")
```

```python
gray_img
```

* Lưu ảnh vào máy:

```python
gray_img.save("new_yasuo.jpg")
```

* Phóng to ảnh:

```python
bigger_img = gray_img.resize((500,500))
```

```python
bigger_img
```

* Thu nhỏ ảnh:

```python
smaller_img = gray_img.resize((100,100))
```

```python
smaller_img
```

* Xoay bức ảnh:

```python
rotate_img = origin_img.rotate(90)
```

```python
rotate_img
```

* Phản chiếu của bức ảnh:

```python
out = origin_img.transpose(PIL.Image.FLIP_LEFT_RIGHT)
out
```
