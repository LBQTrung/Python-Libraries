## Bài tập 2 - Thư viện thao tác với cấu trúc Tensor:
Thực ra thư viện NumPy là thư viện nền tảng,
dựa trên nó có nhiều thư viện được phát triển chuyên biệt. Để thao tác với cấu trúc Tensor có 2
thư viện rất nổi tiếng, gồm:
- TensorFlow
- PyTorch

Em xin phép chọn thư viên PyTorch để tương tác với Tensor

Import thư viện Pytorch:

```python
import torch
```

Tạo hàm show kết quả ra màn hình:

```python
def show_tensor(tensor):
  print("Tensor: \n", tensor)
  print("Tensor hạng: ", tensor.ndim)
  print("Kích thước của tensor: ", tensor.shape)
```

Tạo tensor hạng 1:

```python
data = [1,2,3]
tensor_1 = torch.tensor(data)
```

```python
show_tensor(tensor_1)
```

Tạo tensor hạng 2:

```python
tensor_2 = torch.rand((2,3))
```

```python
show_tensor(tensor_2)
```

Tạo tensor hạng 3:

```python
tensor_3 = torch.rand((2,3,4))
```

```python
show_tensor(tensor_3)
```

Tạo tensor hạng 4:

```python
tensor_4 = torch.rand((2,3,4,5))
```

```python
show_tensor(tensor_4)
```
